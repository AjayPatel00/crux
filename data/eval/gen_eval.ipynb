```python
def preprocess(text):
    # get rid of special characters
    text = ''.join(c for c in text if c.isalnum() or c.isspace())
    return ' '.join(text.split()).lower()

# Load the dataset
with open('company_tickers.json', 'r') as f:
    raw_data = json.load(f)

data = pd.DataFrame(raw_data.values())
data['company_name_processed'] = [preprocess(name) for name in data['title']]
```

```python
companies = list(set(data['company_name_processed']))
```

```python
chunk_size = 50
chunks = [companies[i:i+chunk_size] for i in range(0, len(companies), chunk_size)]
```

```python
def stringify(lst):
    s = ""
    for a in lst:
        s += '\n'+str(a)
    return s[1:]

def api_call(lst):
    prompt_prefix = """
    I need to create an evaluation dataset. I am building a search engine to help users find company names on the stock market. They will be searching for a company, and my search engine will return the best match. The search query may have typos, for example, if the input is amizon, we need to return Amazon Inc. Also if the user inputs google, we need to return Alphabet Inc.
    
    Your task is to generate potential search queries for each of the given company names below. I am expecting output in CSV file format. Each row of your output should be a company name followed by a potential search query, so for example "Alphabet Inc,Google". Another row could be "Alphabet Inc,alphabet". Generate 2 potential search queries per company name. 

    ###Now complete the task I described for the following companies:
    """
    completion = openai.ChatCompletion.create(
    model="gpt-3.5-turbo",
    temperature=0,
    messages=[
        {"role": "user", "content": prompt_prefix + stringify(lst)}
    ])
    return completion.choices[0].message["content"]
```

```python
eval_set = []
```

```python
for i,chunk in enumerate(chunks):
    print("i:", i)
    res = api_call(chunk)
    eval_set.append(res)
    print(res)
```
